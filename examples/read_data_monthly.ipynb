```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Data folder 
data_folder = "/Users/vuongdai/GitHub/bm/detrend_data/monthly/"

data_file = "ts_monthly_values_standardize.csv"
data_file_time = "ts_monthly_datetimes.csv"

# Column to read
cols = [1]
```

```python
sensor_names = pd.read_csv(data_folder + data_file, nrows=0, delimiter=",",  usecols=cols).columns.tolist()
df = pd.read_csv(data_folder + data_file, skiprows=1, delimiter=",", header=None, usecols=cols)
df_time = pd.read_csv(data_folder + data_file_time, skiprows=1, delimiter=",", header=None, usecols=cols)
last_idx = df.index.get_loc(df.iloc[:, 0].last_valid_index())
df = df.iloc[0:last_idx,:]
df_time = df_time.iloc[0:last_idx,:]
```
